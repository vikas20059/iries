```python
import gradio as gr

def frnd_calc(p1, p2):
    import random
    fc = random.randint(0, 100)
    return f"The friendship bonding between {p1} and {p2} is {fc}%"

iface = gr.Interface(
    fn=frnd_calc,
    inputs=[gr.Textbox(label="Your Name"), gr.Textbox(label="Friend Name")],
    outputs=gr.Textbox(label="Bonding Percentage"),
    title="Friendship Calculator"
)

iface.launch()
```
